 # Jigsaw Puzzle 
The goal is to use image processing techniques to solve a jigsaw puzzle.

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
path = 'jigsaw.jpg'
image = cv2.imread(path)
image.shape
```

```python
gsimage = image[:,:,:]
cv2_imshow(gsimage)
```

```python
# convolution filter used to calculate horizontal edges
sobel_x = np.array([[-1, -1, -1], [2, 2, 2],[-1, -1, -1]])
# convolution filter used to calculate verical edges
sobel_y = np.array([[-1, 2, -1], [-1, 2, -1],[-1, 2, -1]])
```

```python
# function to calculate the convolved image
def conv_nested(image, kernel):
    Hi, Wi = image.shape
    Hk, Wk = kernel.shape
    out = np.zeros((Hi, Wi))

    ### YOUR CODE HERE
    for m in range(Hi):
      for n in range(Wi):
        sum=0
        for i in range(Hk):
          for j in range(Wk):
            if m+1-i < 0 or n+1-j < 0 or m+1-i >= Hi or n+1-j >= Wi:
                sum += 0
            else:
                sum += kernel[i][j] * image[m+1-i][n+1-j]
        out[m][n] = sum
    ### END YOUR CODE

    return out
```

```python
# detecting horizontal and vertical edges in the original image
out_h = conv_nested(gsimage, sobel_x)
out_v = conv_nested(gsimage, sobel_y)
```

```python
out_hv = out_h + out_v
cv2_imshow(out_hv)
```

Calculating all the vertices corresponding to the boxes using thresholding.

```python
point = []
for i in range(0,796):
  if(out_hv[10,i]-out_hv[10,i+1]>=200):
    point.append(i)
print(point)
```

```python
point1 = []
for i in range(0,796):
  if(out_hv[300,i]-out_hv[300,i+1]>=200):
    point1.append(i)
print(point1)
```

```python
point2 = []
for i in range(0,420):
  if(out_hv[i,10]-out_hv[i+1,10]>=250):
    point2.append(i)
print(point2)
```

```python
point3 = []
for i in range(0,420):
  if(out_hv[i,530]-out_hv[i+1,530]>=100):
    point3.append(i)
print(point3)
```

```python
point4 = []
for i in range(0,796):
  if(out_hv[410,i]-out_hv[410,i+1]>=150):
    point4.append(i)
print(point4)
```

```python
box1_j1 = 0
box1_j2 = point[1]+1
box1_i1 = 0
box1_i2 = point2[1]
box2_j1 = 0
box2_j2 = point[1]+1
box2_i1 = box1_i2
box2_i2 = point2[2]
box3_j1 = point1[1]+1
box3_j2 = point1[2]
box3_i1 = point3[2]
box3_i2 = point3[3]+1
box4_j1 = point4[2]
box4_j2 = 796
box4_i1 = point3[6]
box4_i2 = 420
```

```python
box1 = np.zeros((box1_i2-box1_i1,box1_j2-box1_j1,3))
for k in range(3):
  for i in range(0,box1_i2):
    for j in range(0,box1_j2):
      box1[i,j,k] = image[i,j,k]
box2 = np.zeros((box2_i2-box2_i1+1,box2_j2-box2_j1+1,3))
for k in range(3):
  for i in range(box2_i1,box2_i2+1):
    for j in range(box2_j1,box2_j2+1):
      box2[i-box2_i1,j-box2_j1,k] = image[i,j,k]
box3 = np.zeros((box3_i2-box3_i1+1,box3_j2-box3_j1+1,3))
for k in range(3):
  for i in range(box3_i1,box3_i2+1):
    for j in range(box3_j1,box3_j2+1):
      box3[i-box3_i1,j-box3_j1,k] = image[i,j,k]
box4 = np.zeros((box4_i2-box4_i1+1,box4_j2-box4_j1+1,3))
for k in range(3):
  for i in range(box4_i1,box4_i2+1):
    for j in range(box4_j1,box4_j2+1):
      box4[i-box4_i1,j-box4_j1,k] = image[i,j,k]
```

```python
box_2 = np.array(box2)
for k in range(3):
  for i in range(box2_i2-box2_i1+1):
    for j in range(box2_j2-box2_j1+1):
      box_2[i,j,k] = box2[box2_i2-box2_i1-i,j,k]
```

```python
box_3 = np.array(box3)
for k in range(3):
  for i in range(box3_i2-box3_i1+1):
    for j in range(box3_j2-box3_j1+1):
      box_3[i,j,k] = box3[i,box3_j2-box3_j1-j,k]
```

```python
box_4 = np.array(box4)
for k in range(3):
  for i in range(box4_i2-box4_i1+1):
    for j in range(box4_j2-box4_j1+1):
      box_4[i,j,k] = box4[box4_i2-box4_i1-i,j,k]
```

```python
def output(path):
  image = cv2.imread(path)
  out_image = image
  for k in range(3):
    for i in range(box3_i1,box3_i2):
      for j in range(box3_j1,box3_j2+1):
        out_image[i,j,k] = box_3[i-box3_i1,j-box3_j1,k]
  for k in range(3):
    for i in range(0,box2_i2-box2_i1+1):
      for j in range(0,box1_j2):
        out_image[i,j,k] = box_2[i,j,k]
  for k in range(3):
    for i in range(0,box1_i2):
      for j in range(0,box1_j2):
        out_image[i+box2_i1,j,k] = box1[i,j,k]
  for k in range(3):
    for i in range(0,box4_i2-box4_i1+1):
      for j in range(0,box4_j2-box4_j1+1):
        out_image[i+box4_i1,j+box4_j1,k] = box_4[i,j,k]
  return out_image
```

```python
def out(path):
  out_image = output(path)
  for i in range(box2_i1,box2_i2):
    for j in range(box2_j1,box2_j2):
      k = out_image[i,j,0]
      out_image[i,j,0] = out_image[i,j,1]
      out_image[i,j,1] = k
  return out_image
```

```python
outp = out(path)
plt.imsave('jigsolved.jpg', outp)
cv2_imshow(outp)
```
